```
# import tensorflow libraries
import tensorflow as tf
from tensorflow import keras  # for sequential stacking of layers
from tensorflow.keras import layers # for making dense layers in neural network

import numpy as np
```

**Define 1-layer Neural Network**

```
model = keras.Sequential([keras.layers.Dense(units=1,input_shape = [1])]) # set up number of layers and neurons
model.compile(optimizer='sgd',loss='mean_squared_error') # model compilation
```

**Array of input variable and output variable**

```
xs = np.array([-3.0,-1.0,0.0,1.0,2.0,3.0,4.0,7.0],dtype=float)
ys = np.array([-25.0,-15.0,-10.0,-5.0,0.0,5.0,10.0,25.0],dtype=float)
```

**Fit the data into the model**

```
model.fit(xs,ys,epochs=5000) # model training
```

**Predict the value on test data**

```
print(model.predict([10.0]))
```
